GROUP BY
- column to group data (categorical)
- output column to display (optional)
- aggregation function (required)
  - 'mean','sum','max','min','std','count'


....
 df.groupby('colname')[col_to_out].sum()
 ...

```python
df.groupby('embark_town').mean()
```

```python
df.groupby('embark_town').sum()
```

```python
df.groupby('embark_town')['age'].mean()
```

```python
df.groupby('embark_town')['fare'].max()
```

```python
df.groupby('embark_town')['survived'].sum()
```

```python
df.groupby('embark_town')['survived'].agg(['sum','mean','std','count'])
```

```python
df.groupby('embark_town')['survived'].agg(['sum','mean','std','count']).T.plot()
```

```python
df.groupby('embark_town')['survived'].agg(['sum']).plot(title='survive')
```

```python
df.groupby('embark_town')['survived'].agg(['sum']).plot(title='Survive',kind='bar')
```

```python
df.groupby('embark_town')['survived'].agg(['sum','count']).plot(title='Survive',kind='barh')
```

```python
df.groupby(['embark_town','who']).sum().style.bar('survived')
```

```python
df.groupby(['embark_town','who'])['survived'].sum().reset_index().style.bar('survived')
```

MAKING AGE AS INTEGER TYPE

```python
def age_roundoff(age):
    try:
        return int(age)
    except:
        return 0    
```

```python
df.age=df.age.apply(age_roundoff)
```
