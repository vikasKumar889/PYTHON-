```python
x = np.arange(-10,10,0.1)
print(x.shape)
```

```python
x2 = x ** 2 # square of x
```

```python
plt.plot(x,x2)
```

```python
x3 = x ** 3
```

```python
plt.figure(figsize=(15,7)) # create a fig with a size of 15*7
plt.plot(x, label='original',linestyle='--')
plt.plot(x2, label='squared',linestyle=':')
plt.plot(x3,label='cubed')
plt.legend()
plt.xlabel('Data points')
plt.ylabel('Values of x')
plt.title('The visualization of X and its square and cube')# add a title to the fig
plt.show() #show the fig
```
